```python
import matplotlib.pyplot as plt
import math
with open('scores.log','r') as f:
    d = list(map(lambda x: float(x), (f.read().split(','))[0:-1]))
        
plt.plot(list(map(lambda x: 100*(1-x),d)))
plt.ylabel('error')
plt.xlabel('epoch')
plt.ylim([0,100])
plt.xlim([0,50])
plt.rcParams['figure.dpi'] = 100
plt.show()
```

```python
plt.plot(list(map(lambda x: 100*x, d)))
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.ylim([0,100])
plt.xlim([0,100])
plt.rcParams['figure.dpi'] = 100
plt.show()
```
